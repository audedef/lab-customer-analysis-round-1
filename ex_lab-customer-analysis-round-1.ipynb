```python
# Read the three files into python as dataframes

import pandas as pd
file1 = pd.read_csv('file1.csv')
file2 = pd.read_csv('file2.csv')
file3 = pd.read_csv('file3.csv')

print(list(file1.columns))
print(list(file2.columns))
print(list(file3.columns))
```

```python
# Show the DataFrame's shape.

print(file1.shape)
print(file2.shape)
print(file3.shape)
```

```python
#Standardize header names.

cols1 = []
for column in file1.columns:
    cols1.append(column.lower())
file1.columns = cols1

cols2 = []
for column in file2.columns:
    cols2.append(column.lower())
file2.columns = cols2

cols3 = []
for column in file3.columns:
    cols3.append(column.lower())
file3.columns = cols3

display(file1.head())
display(file2.head())
display(file3.head())
```

```python
file1.columns = ['customer', 'state', 'gender', 'education', 'cust_lifetime_val', 'income', 'monthly_prem_auto', 'nb_open_complaints', 'policy_type', 'vehicle_class', 'total_claim_amount']
print(file1.columns)
```

```python
file2.columns = ['customer', 'state', 'gender', 'education', 'cust_lifetime_val', 'income', 'monthly_prem_auto', 'nb_open_complaints', 'total_claim_amount', 'policy_type', 'vehicle_class']
display(file2.head())
```

```python
file3.columns = ['customer', 'state', 'cust_lifetime_val', 'education', 'gender', 'income', 'monthly_prem_auto', 'nb_open_complaints', 'policy_type', 'total_claim_amount', 'vehicle_class']
display(file3.head())
```

```python
#Rearrange the columns in the dataframe as needed

file2 = file2[['customer', 'state', 'gender', 'education', 'cust_lifetime_val', 'income', 'monthly_prem_auto', 'nb_open_complaints', 'policy_type', 'vehicle_class', 'total_claim_amount']]
file3 = file3[['customer', 'state', 'gender', 'education', 'cust_lifetime_val', 'income', 'monthly_prem_auto', 'nb_open_complaints', 'policy_type', 'vehicle_class', 'total_claim_amount']]

display(file1.head())
display(file2.head())
display(file3.head())
```

```python
# Concatenate the three dataframes

df = pd.concat([file1, file2, file3], axis = 0)
df = df.reset_index(drop = True)
df.info()
```

```python
# Delete the column education and the number of open complaints from the dataframe.
df = df.drop_duplicates()
df = df.drop(['education', 'nb_open_complaints'], axis = 1)
display(df.head())
```

```python
# Correct the values in the column customer lifetime value. They are given as a percent, so multiply them by 100 and change dtype to numerical type.

df['cust_lifetime_val'] = df['cust_lifetime_val'].str.rstrip('%').astype('float') / 100.0
display(df.head())
```
